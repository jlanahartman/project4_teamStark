```python
import psycopg2
import pandas as pd
```

```python
# Establish a connection to the PostgreSQL database
conn = psycopg2.connect(database="project4_teamStark", user="postgres", password="password123", host="127.0.0.1", port="5432")

# Create a cursor object
cursor = conn.cursor()

# Execute a SQL query
cursor.execute("SELECT * FROM winedf")

# Fetch all the rows
rows = cursor.fetchall()

# Convert the rows to a pandas DataFrame
df = pd.DataFrame(rows)

# Close the cursor and connection
cursor.close()
conn.close()
```
